Conversion to GrayScale



```python
#Import required library
from PIL import Image

# Visualization library
import matplotlib.pyplot as plt

#Open/Load image using 'Image' module
#Provide both name and path to the image file
imLena = Image.open('lena.png')

# Convert image to gray scale
imLenaGray = imLena.convert('L')
# display( imLenaGray )

# Creates a plot with 1 row and 3 columns and display in the first position
plt.figure( figsize = (15,15))
plt.subplot(1, 3, 1)
plt.imshow( imLena )

# Use the second place to display the grayscale image
plt.subplot(1, 3, 2)
plt.imshow( imLenaGray )

# Use the third place to display the grayscale image in gray map
plt.subplot(1, 3, 3)
plt.imshow( imLenaGray, cmap = 'gray' )
```

Now, let's split the image into its R, G, B components:

```python
# Splitting the image into individual color bands
imLenaR, imLenaG, imLenaB, = imLena.split()

# Creates a plot 
plt.figure( figsize = (15,15))

# Plotting each of these channels shows each channel as a Gray Scale Image
# Display the Red channel
plt.subplot(1, 3, 1)
plt.imshow( imLenaR )

# Display the Green channel
plt.subplot(1, 3, 2)
plt.imshow( imLenaG )

# Display the Blue channel
plt.subplot(1, 3, 3)
plt.imshow( imLenaB )
```

```python
# However, if we create an RGB image with only values in one channel and others
# zero, then we get can display the true, R, G, B components

# Another library - provides comprehensive mathematical functions and operations
import numpy as np
   
# Creates a plot 
plt.figure( figsize = (15,15))

# Creat a Zero matrix with the same shape as original image
# Then, assigne one channel at a time
imZeroR = np.zeros( (512, 512, 3), dtype="uint8" ) # 'dtype' by default: 'numpy.float64'
imZeroR[ :, :, 0] = imLenaR

imZeroG = np.zeros( (512, 512, 3), dtype="uint8" ) # 'dtype' by default: 'numpy.float64'
imZeroG[ :, :, 1] = imLenaG

imZeroB = np.zeros( (512, 512, 3), dtype="uint8" ) # 'dtype' by default: 'numpy.float64'
imZeroB[ :, :, 2] = imLenaB

plt.subplot(1, 3, 1)
plt.imshow( imZeroR )

plt.subplot(1, 3, 2)
plt.imshow( imZeroG )

plt.subplot(1, 3, 3)
plt.imshow( imZeroB )
```

Let's try manipulating the individual color bands.

As an example, let’s set the Red, Green, Blue layers for following Rows values to full intensity.

R channel: Row- 100 to 110
G channel: Row- 200 to 210
B channel: Row- 300 to 310

```python
# Another library - provides comprehensive mathematical functions and operations
import numpy as np

# First let's make a copies of our original image
# Also, image objects cannot be manipulated directly
# So, first we conver it into a simple array/matrix
cpLenaR = np.array( imLena, np.uint8 )
cpLenaG = np.array( imLena, np.uint8 )
cpLenaB = np.array( imLena, np.uint8 )

# Create a figure for displaying
plt.figure( figsize = (15,15))

# Set max intensity to pixel's in rows 50:150 in R channel
cpLenaR[50:150 , : , 0] = 255 
plt.subplot(1,3,1)
plt.imshow( cpLenaR )

# Set max intensity to pixel's in rows 200:300 in G channel
cpLenaG[200:300 , : , 1] = 255
plt.subplot(1,3,2)
plt.imshow( cpLenaG )

cpLenaB[350:450 , : , 2] = 255 # full intensity to those pixel's B channel
plt.subplot(1,3,3)
plt.imshow( cpLenaB )

plt.show()
```
